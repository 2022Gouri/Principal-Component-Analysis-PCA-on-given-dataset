```python
arr_standard=(arr-np.mean(arr,axis=0))
print("Standardized data\n",arr_standard)
```

```python
cov_mat=np.cov(arr_standard,rowvar=False)
print("Covariance matrix\n",cov_mat)
```

```python
eigen_values,eigen_vectors=np.linalg.eig(cov_mat)
```

```python
sorted_index = np.argsort(eigen_values)[::-1]
sorted_eigenvalue = eigen_values[sorted_index]
print("Eigenvalues sorted")
print(sorted_eigenvalue)
sorted_eigenvectors = eigen_vectors[:,sorted_index]
print("Eigenvectors sorted")
print(sorted_eigenvectors)
```

```python
eigenvector_subset = sorted_eigenvectors[:,0:2]
print("Top 2 Eigen vectors")
print(eigenvector_subset)
```

```python
arr_reduced = np.dot(arr_standard,eigenvector_subset)
print("New transformed data with two columns pc1 and pc2")
print(arr_reduced)
```

```python
arr_reduced = np.dot(arr_standard,eigenvector_subset)
print("New transformed data with two columns pc1 and pc2")
print(arr_reduced)

```

```python
arr_reduced = np.dot(arr_standard,eigenvector_subset)
print("New transformed data with two columns pc1 and pc2")
print(arr_reduced)
#Plot the reduced data
for pc1, pc2, lab, col in zip(arr_reduced.transpose()[0],arr_reduced.transpose()[1],('brand','model','base_color','ROM','RAM','display_size','num_rear_camera','num_front_camera','battery_capacity','ratings','num_of_ratings','sales_price','discount_percent','sales'),('blue','red','green','yellow','purple','brown','black','orange','pink','blue','red','green','yellow','orange')):
 plt.scatter(pc1,pc2,c=col)
 plt.annotate(lab,(pc1,pc2))
plt.xlabel("Principal Component 1")
plt.ylabel("Principal Component 2")
plt.show()
```
